```python
#1. Import the libraries required for exploratory data analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#2. We will import our dataset
df=pd.read_csv('SampleSuperstore.csv')
```

```python
df.info()
```

```python
# we will print summary of our dataset
df.describe()
```

```python
#4.we will print first five enteries of our data
df.head()
```

```python
#5.we will check if there are any NAN values in our data
df.isnull().sum()
```

```python
df.shape
```

```python
#6. We will print the column names of our data
df.columns
```

```python
#7. We will calculate sales of company
sale = df['Sales'].sum()
sale
```

```python
#8. We will calculate profit of company
profit = df['Profit'].sum()
profit
```

```python
#9. We will calculate discount of company
discount =df['Discount'].sum()
discount
```

```python
# We will calculate sum of top ten sales of the company
Top_10_sales= df.groupby("State").Sales.sum().nlargest(n=10)
Top_10_sales
```

```python
# we will plot graph of top 10 Sales of company
plt.style.use('grayscale')
Top_10_sales.plot(kind ='bar',color="purple",figsize =(8,6), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title('Top 10 Sales per State',fontsize =16)
plt.show()
```

```python
# We will calculate sum of top ten Profit of the company
Top_10_profit = df.groupby("State").Profit.sum().nlargest(n=10)
Top_10_profit
```

```python
# we will plot graph of top 10 Profit of company
Top_10_profit.plot(kind ='bar',color="purple", figsize =(10,6), fontsize =14)
plt.xlabel("States", fontsize =14)
plt.ylabel("Total Profit",fontsize =14)
plt.title("Top 10 Profits per State",fontsize =16)
plt.show()
```

```python
# we will find the correlation between entries in our data
df.corr()
```

```python
# We will plot the correlation graph
sns.heatmap(df.corr(),annot=True,cmap='BuPu')
```

```python
# We will make count on the basis of category
df2=df['Category'].value_counts()
```

```python
df2
```

```python
# we will plot countplot on the basis of category
sns.countplot(x='Category',data=df,palette='bright')
```

```python
# We will make count on the basis of Sub-category
df['Sub-Category'].value_counts()
```

```python
# we will plot countplot on the basis of Sub-category
plt.figure(figsize=(12,6))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
# We will calculate the sum of Profit based on Category
df.groupby(['Category'])['Profit'].sum()
```

```python
# We will plot barplot of Category in terms of Profit
sns.barplot(x='Category',y='Profit',data=df,palette='bright')
```

```python
# We will calculate the sum of Discount based on Category
df.groupby(['Category'])['Discount'].sum()
```

```python
# We will plot barplot of Category in terms of Discount
sns.barplot(x='Category',y='Discount',data=df,palette='bright')
```

```python
# We will calculate the sum of Sales based on Category
df.groupby(['Category'])['Sales'].sum()
```

```python
# We will plot barplot of Category in terms of Sales
sns.barplot(x='Category',y='Sales',data=df,palette='bright')
```

```python
# We will plot Pairplot on the basis of Category
sns.pairplot(df,hue='Category',palette='mako')
```

```python
#We will calculate sum,mean,count, median,standard deviation,Variance of each states of Profit


df.groupby("State").Profit.agg(["sum","mean",'median','std','count','var'])
```

```python
# We will plot countplot of Sub-Categoreis as per the Region
plt.figure(figsize = (10, 8))
sns.countplot(x = 'Sub-Category', hue = 'Region', data = df, palette ='viridis')
plt.xticks(rotation = 90, fontsize = 11)
plt.yticks(fontsize = 11)
plt.xlabel('Sub-Categories',fontsize = 20)
plt.legend(loc = 5, fontsize = 10)
plt.show()
```

```python
# We will plot countplot of Sub-Categoreis as per the Region
plt.figure(figsize = (10, 8))
sns.countplot(x = 'Category', hue = 'Region', data = df, palette = 'viridis')
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Categories',fontsize = 20)
plt.legend(loc = 5, fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(x = df["Segment"],palette='BuPu')
```

```python
# we will plot countplot to Visualize the State
plt.figure(figsize=(20,8))
sns.countplot(x=df['State'])
plt.xticks(rotation=90,fontsize = 14)
plt.yticks(fontsize=14)
plt.show()
```

```python
#We will plot barplot for Sub-Category and Sales
plt.figure(figsize=[20,8])
ax = sns.barplot(x="Sub-Category", y="Sales", data=df, palette="Blues",edgecolor='black')
plt.xlabel("Subcategories",fontsize=15)
plt.ylabel("Sales",fontsize=15)
```

```python
#We will plot barplot for Sub-Category and Profit
plt.figure(figsize=[10,8])
ax = sns.barplot(x="Sub-Category", y="Profit", data=df, palette="mako",edgecolor='black')
plt.xlabel("Subcategory", fontsize=15)
plt.ylabel("Profit",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
# we will plot Barplot for the Profit Per State
plt.figure(figsize=[20,15])
ax = sns.barplot(x="State", y="Profit", data=df, palette="viridis",edgecolor='black')
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=15)
plt.title("Profit per State",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("Profit",fontsize=20)
plt.tight_layout()
```
